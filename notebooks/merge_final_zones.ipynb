# merge all zones real distances in a csv

```python
import pandas as pd

df1 = pd.read_csv("/Users/anuragchaubey/Desktop/data/northern_zone.csv")
df2 = pd.read_csv("/Users/anuragchaubey/Desktop/data/central_zone.csv")
df3 = pd.read_csv("/Users/anuragchaubey/Desktop/data/southern_zone.csv")    
df4 = pd.read_csv("/Users/anuragchaubey/Desktop/data/eastern_zone.csv")
df5 = pd.read_csv("/Users/anuragchaubey/Desktop/data/western_zone.csv")
df6 = pd.read_csv("/Users/anuragchaubey/Desktop/data/north_eastern_zone.csv")
```

```python
merged_df = df1.copy()

# list of other dataframes
dfs = [df2, df3, df4, df5, df6]

# fill missing distances from each zone's data
for df in dfs:
    merged_df["osrm_road_distance_km"] = merged_df["osrm_road_distance_km"].where(
        merged_df["osrm_road_distance_km"] != 0,  # keep if already non-zero
        df["osrm_road_distance_km"]               # otherwise take from this df
    )

```

```python
non_zero_count = (merged_df["osrm_road_distance_km"] != 0).sum()

print(f" Non-zero distance count: {non_zero_count}")
merged_df.shape
```

```python
df = merged_df[merged_df["osrm_road_distance_km"] != 0] 

df.shape
```

```python
df.columns = ["real_distance_km" if col == "osrm_road_distance_km" else col for col in df.columns]
```

```python
# Save final CSV
df.to_csv("/Users/anuragchaubey/MilesAhead/data/processed/delivery_data_real_distance.csv", index=False)
```

## convert all columns to appropriate data types

```python
# convert 'Order_Date' to datetime
df_final['Order_Date'] = pd.to_datetime(df_final['Order_Date'], errors='coerce')
```

```python
# Add dummy date to convert time strings to datetime 

time_format = "%Y-%m-%d %H:%M:%S"

df_final['Order_Time'] = pd.to_datetime(
    "1900-01-01 " + df_final['Order_Time'].astype(str),
    format=time_format,
    errors='coerce'
)

df_final['Pickup_Time'] = pd.to_datetime(
    "1900-01-01 " + df_final['Pickup_Time'].astype(str),
    format=time_format,
    errors='coerce'
)
```
